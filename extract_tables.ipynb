# Cleaning

```python
def get_empty_rows(dataset):
    # Nested 'for' loop and unique value list of 'set'
    return list(set([i for i, s in dataset.iterrows() for v in s.values if str(v) == 'nan']))

import pandas as pd

# for table in tables:
#     df = table.df
#     break

year = "2021"


# Declarar cabecera
# header = df.iloc[0]
header = ['NRO'
          ,'RUC'
          ,'NOMBRE O RAZON SOCIAL'
          ,'APORTE TOTAL(mm)'
        ]

# create a empty dataframe
df_contenedor = pd.DataFrame(columns=header)

# iterar por cada tabla del PDF
for table in tables:
    df = table.df
    # remove the first row with loc
    df = df.loc[1:]
    df.columns = header
    df['ANHO'] = year


    # remove rows with empty string based on NRO column with loc
    df = df.loc[df['NRO'] != '']
    # df = df.loc[df['NRO'].notnull()]
    # df = df.dropna(subset=['Nro'])

    # concatenate the dataframe into a container dataframe
    df_contenedor = pd.concat([df_contenedor, df], ignore_index=True)

    list_of_empty_rows = get_empty_rows(df)

    # year -= 1


    # break
    # print(df)
    # Guardar en CSV
    # df.to_csv('csvs/csv1.csv', mode='a', header=True, index=False)


# limpiar lineas innecesarias


list_of_empty_rows
```
